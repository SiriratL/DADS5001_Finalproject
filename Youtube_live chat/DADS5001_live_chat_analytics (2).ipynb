```python
import pytchat
chat = pytchat.create(video_id = "Fncg9oH16O0")

datelst = []
namelst = []
commentlst = []

while chat.is_alive():
  for c in chat.get().sync_items():
    print(f"{c.datetime} [{c.author.name}]- {c.message}")
    datelst.append(c.datetime)
    namelst.append(c.author.name)
    commentlst.append(c.message)
```

```python
#dictionary of list
dict = {'DateTime': datelst, 'Username': namelst, 'Comment': commentlst}
```

```python
import pandas as pd
df = pd.DataFrame(dict)
df.head()
```

```python
from google.colab import files
df.to_csv("data.csv", encoding="utf_8_sig", index=False) 
#cp874, 
#TIS-620
files.download("data.csv")
```
